```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
T = 1000

B_start = 0
B_end = 1

x = np.linspace(0,T,1001)
y = np.empty_like(x)
```

```python
# Lineal function
def Lineal(x: np.ndarray, T:int):
    y = np.empty_like(x)
    for t in range(T+1):
        y[t] = B_start + (B_end-B_start)*(t/T)

    return y
```

```python
plt.grid(linewidth=0.3)
plt.plot(x,Lineal(x,T))

plt.title("Planificador Lineal")
plt.xlabel("t")
plt.ylabel(r"$\beta$")

plt.yticks([0,0.2,0.4,0.6,0.8,1],[r"$\beta_{start}$",0.2,0.4,0.6,0.8,r"$\beta_{end}$"]);
```

```python
#Cosenoidal
def cosenoidal(x:np.ndarray, T:int):
    y = np.empty_like(x)
    for t in range(T+1):
        y[t] = B_end + 0.5*(B_start-B_end)*(1+np.cos(np.pi*t/T))

    return y

plt.plot(x,cosenoidal(x,T))
plt.grid(linewidth=0.3)
plt.title("Planificador Cosenoidal")

plt.xlabel("t")
plt.ylabel(r"$\beta$")

plt.yticks([0,0.2,0.4,0.6,0.8,1],[r"$\beta_{start}$",0.2,0.4,0.6,0.8,r"$\beta_{end}$"]);
```

```python
#exp
B_start = 0.01
B_end = 1

def exponential(x:np.ndarray, T:int):
    y = np.empty_like(x)
    for t in range(T+1):
        y[t] = B_start * (B_end/B_start)**(t/T)

    return y

plt.plot(x,exponential(x,T))
plt.grid(linewidth=0.3)
plt.title("Planificador exponencial")

plt.xlabel("t")
plt.ylabel(r"$\beta$")

plt.yticks([0.01,0.2,0.4,0.6,0.8,1],[r"$\beta_{start}$",0.2,0.4,0.6,0.8,r"$\beta_{end}$"]);
```
